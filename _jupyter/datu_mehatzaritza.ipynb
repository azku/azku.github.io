Hemen berehala ikusten ditugu zenbat datu erregistro ditugun (21613) eta etxean prezioan mediak zeintzuk diren.

Hurrengo pausoan grafika batzuk aterako ditugu matplotlib (plt) liburutegia erabilita.Kasu honetan "square foot" edo etxebizitzaren tamainua adierazten duen aldagai bat eta etxebizitzen prezioaren aldagaia ateratzen ditugu. Kasu bietan ezker aldean daude kasu gehienak.
Gogoratu histograman x ardatzean aldagaia aterako dugula eta y ardatzean balio hori duten erregistro kopurua.

```python
fig = plt.figure(figsize=(12, 6))
sqft = fig.add_subplot(121)
cost = fig.add_subplot(122)

sqft.hist(df.sqft_living, bins=80)
sqft.set_xlabel('Ft^2')
sqft.set_title("Histogram of House Square Footage")

cost.hist(df.price, bins=80)
cost.set_xlabel('Price ($)')
cost.set_title("Histogram of Housing Prices")
```

Gure datu-multzoa ondo ezagutzen dugunez eta neurtu nahi ditugun aldagaien banaketak ezagutzen ditugunez, egin dezagun erregresio-analisi bat. Lehenik, statsmodels inportatuko dugu erregresio funtzio karratu txikiena lortzeko. "ordinary least squares" moduluak egingo du lanaren zatirik handiena.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

2 aldagaien regresio linearra burutu nahi badugu ondorengo formaula erabiliko dugu:

    Reg = ols(‘Dependent variable ~ independent variable(s), dataframe).fit()
    print(Reg.summary())
Hau python kodera eramaten badugu

```python
m = ols('price ~ sqft_living',df).fit()
print (m.summary())
```

Eta ondotengo kodea?

```python
m = ols('price ~ sqft_living + bedrooms + grade + condition',df).fit()
print (m.summary())
```

```python
sns.jointplot(x="sqft_living", y="price", data=df, kind = 'reg',fit_reg= True)
plt.show()
```
